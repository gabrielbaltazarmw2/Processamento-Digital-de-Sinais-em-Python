#Função Impulso

```python
# função impulso como uma função lambda
imp_function = lambda t: np.where(np.logical_and(-0.1e-8 < t, t < 0.1e-8), 1, 0)
```

```python
# vetor de tempo
t = np.arange(-4,10,0.01)
```

```python
t
```

```python
# função impulso delta
delta = imp_function
```

```python
# Plot
plt.figure(figsize=(8,3))
plt.plot(t, delta(t), label='$\delta$(t)')
plt.xlabel('Tempo', fontsize=20)
plt.ylabel('Valor')
plt.title('Função Impulso $\delta$(t)')
plt.ylim(-0.2,1.2)
plt.xlim(-4,10)
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

```python
# Plot 2
plt.figure(figsize=(8,3))
plt.plot(t, 2*delta(t-3), label='$2\delta(t-3)$', )
plt.xlabel('Tempo', fontsize=20)
plt.ylabel('Valor')
plt.title('Função Impulso $2\delta(t-3)$')
plt.ylim(top=2.2)
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

```python
#plot 3
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.plot(t, delta(t), label='$\delta$(t)')
plt.xlabel('Tempo', fontsize=20)
plt.ylabel('Valor')
plt.title('Função Impulso $\delta$(t)')
plt.ylim(top=5.2)
plt.legend(loc='upper right')
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(t, 2*delta(t-3), label='$2\delta(t-3)$', )
plt.xlabel('Tempo', fontsize=20)
plt.ylabel('Valor')
plt.title('Função Impulso $2\delta(t-3)$')
plt.ylim(top=5.2)
plt.legend(loc='upper right')
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(t, 0.5*delta(t+2), label='$2\delta(t-3)$', )
plt.xlabel('Tempo', fontsize=20)
plt.ylabel('Valor')
plt.title('Função Impulso $0.5\delta(t+2)$')
plt.ylim(top=5.2)
plt.legend(loc='upper right')
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(t, 5*delta(t-5), label='$5\delta(t-5)$', )
plt.xlabel('Tempo', fontsize=20)
plt.ylabel('Valor')
plt.title('Função Impulso $5\delta(t-5)$')
plt.ylim(top=5.2)
plt.legend(loc='upper left', fontsize=10)
plt.grid(True)

# Ajuste de espaçamento entre os subplots
plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```

```python
def subplots_impulsos(impulso1,impulso2,impulso3,impulso4):

  plt.figure(figsize=(10,4))

  plt.subplot(2,2,1)
  plt.plot(t, impulso1)
  plt.xlabel('Tempo', fontsize=10)
  plt.ylabel('Valor')
  plt.title('Impulso 1')
  plt.grid(True)

  plt.subplot(2,2,2)
  plt.plot(t, impulso2)
  plt.xlabel('Tempo', fontsize=10)
  plt.ylabel('Valor')
  plt.title('Impulso 2')
  plt.grid(True)

  plt.subplot(2,2,3)
  plt.plot(t, impulso3)
  plt.xlabel('Tempo', fontsize=10)
  plt.ylabel('Valor')
  plt.title('$Impulso 3$')
  plt.grid(True)

  plt.subplot(2,2,4)
  plt.plot(t, impulso4)
  plt.xlabel('Tempo', fontsize=10)
  plt.ylabel('Valor')
  plt.title('Impulso $4$')
  plt.grid(True)

  # Ajuste de espaçamento entre os subplots
  plt.subplots_adjust(hspace=1, wspace=0.5)
  plt.show()
```

```python
subplots_impulsos(delta(t),delta(t-3),delta(t+4),5*delta(t-5))
```

#Trem de Impulsos Periódicos Definitivo

```python
# função impulso como uma função lambda
delta = lambda n: np.where(np.logical_and(-0.1e-8 < n, n < 0.1e-8), 1, 0)


f=10;                  # frequência do impulso
fs=4*f;                # frequência de amostragem com um fator de oversampling
Ts=1/fs;               # intervalo de amostragem
t= np.arange(0,2,Ts);  # intervalo de tempo do trem de impulso
trem = delta(t)

# # inicializando o vetor trem de impulsos com um único periodo
trem = np.zeros_like(t)  # trem de impulso preenchido do tamanho do vetor de tempo

# # construção do trem de impulsos com impulsos deslocados
for i in range(int(np.floor(np.min(t)))*f,int(np.ceil(np.max(t)))*f):
  trem += (delta(t-(i/f)))

# # plot
plt.stem(t,trem)
```
